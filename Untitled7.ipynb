```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
dataset = pd.read_csv('NKE_quarterly_financial_data.csv')
dataset.shape
dataset.describe()
#dataset.Revenue = dataset.Revenue.astype(np.int64)
print(dataset[['Revenue']])
```

```python
dataset.Revenue = pd.to_numeric(dataset.Revenue, errors= 'coerce')
#dataset.Revenue.apply(lambda x: convert_float(x))
dataset.Revenue.astype(float)
print (dataset[['Revenue']])
```

```python

#dataset.Revenue.astype(float)
dataset.plot(x='Current Liabilities', y='Revenue', style='o')  
plt.title('Revenue vs Current Liabilities')  
plt.xlabel('Revenue')  
plt.ylabel('Current Liabilities')  
plt.show()
```

```python
dataset = dataset.dropna(subset=['Revenue'])
```

```python
dataset.Revenue.astype(int)
plt.figure(figsize=(15,10))
plt.tight_layout()
seabornInstance.distplot(dataset['Revenue'], kde= False, bins = 20)
```

```python
corr = dataset.corr()
ax = seabornInstance.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=seabornInstance.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=90,
    horizontalalignment='right'
);
```

```python
import seaborn as sns
Var_Corr = dataset.corr()
plt.figure(figsize=(5,5))
sns.heatmap(Var_Corr, xticklabels=Var_Corr.columns, yticklabels=Var_Corr.columns, annot=True)
```

```python
dataset['date'] = pd.to_datetime(dataset['Quarter end'],format='%Y-%m-%d')
dataset.index = dataset['date']
```

```python
df = dataset[['Shares', 'Shares split adjusted', 'Assets',
               'Current Assets', 'Liabilities', 'Current Liabilities', 'Shareholders equity','Price','Price high','Price low','Cash at end of period','Book value of equity per share','Cumulative dividends per share','Revenue']]
train = df[20:]
test = df[:20]



X_train = train.drop('Revenue', axis=1)
y_train = train['Revenue']
X_test = test.drop('Revenue', axis=1)
y_test = test['Revenue']

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

lm = LinearRegression()
lm.fit(X_train,y_train)

preds = lm.predict(X_test)
rms=np.sqrt(np.mean(np.power((np.array(y_test)-np.array(preds)),2)))
print(rms)
```

```python

pd.concat([y_test,pd.DataFrame(preds,index = y_test.index)],axis=1).plot()
```

```python
#knn-neighbour
from sklearn import neighbors
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
x_train_scaled = scaler.fit_transform(X_train)
x_train = pd.DataFrame(x_train_scaled)
x_valid_scaled = scaler.fit_transform(X_test)
x_valid = pd.DataFrame(x_valid_scaled)

params = {'n_neighbors':[2,3,4,5,6,7,8,9]}
knn = neighbors.KNeighborsRegressor()
model = GridSearchCV(knn, params, cv=5)

model.fit(x_train,y_train)
preds1 = model.predict(x_valid)

print(preds1)

rms=np.sqrt(np.mean(np.power((np.array(y_test)-np.array(preds1)),2)))

print(rms)

pd.concat([y_test,pd.DataFrame(preds1,index = y_test.index)],axis=1).plot()
```
